Handling missing data

```python
df.dropna(subset=['Customer_ID'], inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Age'].unique()
```

```python
import re

def extract_age(age):
  age_num = re.findall('[0-9]+', str(age))
  if len(age_num)>0:
    return age_num[0]
  else:
    return age

df['Age'] = df['Age'].apply(lambda x : extract_age(x))
```

```python
df
```

```python
df['Age'].unique()
```

```python
df_age = df[df['Age'] != 'nan years'] ['Age']
```

```python
df_age
```

```python
age_median = int(df_age.dropna().astype('int64').median())
```

```python
age_median
```

```python
df.replace('nan years', age_median, inplace = True)
```

```python
df['Age'].fillna(age_median, inplace = True)
```

```python
df['Age'].unique()
```

```python
df['Purchase_Amount'].fillna(df['Purchase_Amount'].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df
```

```python
df['Feedback_Score'].fillna(df['Feedback_Score'].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
for col in ['Gender', 'City', 'Country']:
  df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Last_Purchase_Date'].ffill(inplace=True)
```
